```python
def scrapping(terms):
    al=[]
    for i in terms:
        response=youtube.search().list(q=i,maxResults=50,type="video",part="snippet").execute()
        al += response['items']
    n=len(terms)
    for i in range(n):
        for j in range((i*50), 2500,50):
            for k in range(50):
                al[j:j+50][k]["snippet"]["keyword"] = terms[i]
    # We make the dictionary of video ids keys with video title from 'a' varible
    videos = {}
    for search_result in al:
        if search_result["id"]["kind"] == "youtube#video":
            videos[search_result["id"]["videoId"]] = (
                search_result["snippet"]["title"],search_result["snippet"]["keyword"]
            )
    #We store the all key ids in list format
    video_ids=list(videos.keys())
    #give the video ids in 40-40 batch & fetch the response of all videos statistics,details and will store in the "stats" variable
    stats=[]
    for i in range(0,(len(video_ids)),40):
        res=youtube.videos().list(id=','.join(video_ids[i:i+40]),part='statistics,snippet,id').execute()
        stats+=res['items']
    #We make the dictionary of required variables & store in "res" variable
    res = []
    for i in stats:
        temp_res = dict(v_id = i['id'], v_title = videos[i['id']][0],Keyword=videos[i['id']][1],
                        publishedDate=i['snippet']['publishedAt'],v_description=i['snippet']['description'])
        temp_res.update(i['statistics'])
        res.append(temp_res)
    import pandas as pd
    #Convert the the "res" list to dataframe & save the csv format
    df = pd.DataFrame.from_dict(res)
    df.to_csv('serchresult.csv', mode='a',encoding='utf-8')
```
